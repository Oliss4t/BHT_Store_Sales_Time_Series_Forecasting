```python
def get_sales_by_store_family(test, proportion):
    result = []
    for index, row in test.iterrows():
        store_nbr = row['store_nbr']
        family = row['family']
        sales = proportion.loc[(proportion['store_nbr'] == store_nbr) & (proportion['family'] == family), 'sales'].values[0]
        result.append({'id': row['id'], 'sales': sales})
    return pd.DataFrame(result)
```

```python
def get_predicted_sales(test, proportion, predictions):
    result = []
    for index, row in test.iterrows():
        store_nbr = row['store_nbr']
        family = row['family']
        date = row['date']
        base_sales = proportion.loc[(proportion['store_nbr'] == store_nbr) & (proportion['family'] == family), 'sales'].values[0]
        total_sales = predictions.loc[predictions['date'] == date, 'total_sales'].values[0]
        final_sales = base_sales * total_sales
        result.append({'id': row['id'], 'sales': final_sales})
    return pd.DataFrame(result)
```
